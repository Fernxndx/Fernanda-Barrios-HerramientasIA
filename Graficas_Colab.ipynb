```python
import matplotlib.pyplot as plt
import numpy as np

# Para el primer ejercicio voy a hacer la relacion entre las horas de estudio y puntajes de un examen
np.random.seed(42)
horas_estudio = np.random.rand(50) * 10  # De 0 a 10 horas
puntaje = 2 * horas_estudio + np.random.randn(50) * 2  # Relación lineal con ruido

plt.figure(figsize=(8, 5))
plt.scatter(horas_estudio, puntaje, color='skyblue', edgecolor='black', alpha=0.8)

plt.title('Relación entre Horas de Estudio y Puntaje')
plt.xlabel('Horas de Estudio')
plt.ylabel('Puntaje del Examen')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Para el segundo, nos basaremos en la estatura de 1000 personas donde: (Media 170cm, Desviación Estándar 10)
np.random.seed(42)
estaturas = np.random.normal(170, 10, 1000)

plt.figure(figsize=(8, 5))
plt.hist(estaturas, bins=30, color='salmon', edgecolor='white')

plt.title('Distribución de Estaturas en una Población')
plt.xlabel('Estatura (cm)')
plt.ylabel('Número de Personas')
plt.axvline(np.mean(estaturas), color='blue', linestyle='dashed', linewidth=1, label='Promedio')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de los helados
sabores = ['Vainilla', 'Chocolate', 'Fresa', 'Menta', 'Arequipe']
np.random.seed(42)
ventas = np.random.randint(10, 100, size=5)

plt.figure(figsize=(8, 5))
colores = ['#FFFACD', '#4B2C20', '#FFB6C1', '#98FB98', '#CD853F'] # Colores según el sabor

plt.bar(sabores, ventas, color=colores, edgecolor='gray')

plt.title('Preferencia de Sabores de Helado')
plt.xlabel('Sabores')
plt.ylabel('Unidades Vendidas')
plt.show()
```
